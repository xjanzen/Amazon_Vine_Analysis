```python
import pandas as pd
import os
```

```python
df_to_load = os.path.join("Resources", "vine_table.csv")
df = pd.read_csv(df_to_load)
df.head()
```

```python
df_total_votes = df[(df.total_votes>=20)]
df_total_votes.head()
```

```python
df_helpful_reviews = df_total_votes[(df_total_votes.helpful_votes/df_total_votes.total_votes>=.5)]
df_helpful_reviews.head()
```

```python
df_vine_paid = df_helpful_reviews[(df_helpful_reviews.vine=="Y")]
df_vine_paid.head()
```

```python
df_vine_unpaid = df_helpful_reviews[(df_helpful_reviews.vine=="N")]
df_vine_unpaid.head()
```

```python
paid_reviews = df_vine_paid['review_id'].count()
paid_5star_reviews = df_vine_paid[(df_vine_paid.star_rating==5)]['star_rating'].count()
paid_5star_reviews_percentage = paid_5star_reviews/paid_reviews*100
print('Number of Paid Reviews: '+str(paid_reviews))
print('Number of 5-star Paid Reviews: '+str(paid_5star_reviews))
print('Percentage of 5-star Paid Reviews: '+str(paid_5star_reviews_percentage.round(2))+'%')
```

```python
unpaid_reviews = df_vine_unpaid['review_id'].count()
unpaid_5star_reviews = df_vine_unpaid[(df_vine_unpaid.star_rating==5)]['star_rating'].count()
unpaid_5star_reviews_percentage = unpaid_5star_reviews/unpaid_reviews*100
print('Number of Unpaid Reviews: '+str(unpaid_reviews))
print('Number of 5-star Unpaid Reviews: '+str(unpaid_5star_reviews))
print('Percentage of 5-star unpaid Reviews: '+str(unpaid_5star_reviews_percentage.round(2))+'%')
```
